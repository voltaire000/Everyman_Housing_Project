```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = 'train.csv' 
houses = pd.read_csv('./datasets/' + data)
```

```python
# Making the column names a little more approachable
houses.columns = [x.lower() for x in houses.columns]
houses.columns = [x.replace(' ', '_') for x in houses.columns]
```

```python
houses = houses.fillna(0)
```

```python
plt.figure(figsize = [9, 6])
plt.scatter(houses['saleprice'], houses['lot_area'], c = 'purple')
```

```python
houses = houses[houses.lot_area <= 60000] # handles some of the larger outliers that may affect our data
```

```python
plt.figure(figsize = [9, 6])
plt.scatter(houses['saleprice'], houses['lot_area'], c = 'purple')
```
